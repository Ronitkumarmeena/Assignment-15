```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("tip_data.csv")
df.head()
```

```python
df.tail()
```

```python
len(df)
```

```python
len(df.columns)
```

```python
day_counts = df['day'].value_counts()
plt.bar(day_counts.index, day_counts.values)
plt.xlabel('Day')
plt.ylabel('Number of Entries')
plt.title('Number of Entries for Each Day')
plt.show()
```

```python
plt.scatter(df['total_bill'], df['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Relationship between Total Bill and Tip')
plt.show()
```

```python
plt.hist(df['tip'], bins=10, alpha=0.5, label='Tip')
plt.hist(df['total_bill'], bins=10, alpha=0.5, label='Total Bill')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Tip and Total Bill')
plt.legend()
plt.show()
```

```python
total_sales = df.groupby('day')['total_bill'].sum()
plt.bar(total_sales.index, total_sales.values)
plt.xlabel('Day')
plt.ylabel('Total Price')
plt.title('Total Sales for Each Day')
plt.show()
```

```python
smoker_counts = df['smoker'].value_counts()
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%')
plt.title('Percentage of Smokers')
plt.show()
```
